```python
#intentamos hacer una gráfica de presión contra altura con los datos de Junio
dfj = df1[['HGHT','PRES']]
plt.figure(); dfj.plot(x='HGHT'); plt.legend(loc='best')
plt.title("Presión contra Altura (22/Junio)")
plt.xlabel("Presion (hPa)")
plt.ylabel("Altura(m)")
plt.grid(True)
plt.show()
```
